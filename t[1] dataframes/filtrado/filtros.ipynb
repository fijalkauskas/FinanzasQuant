selección por filtros condicionales

```python
filtro = aapl[aapl['volume'] > 50000000]
filtro
```

filtros avanzados

```python
filtro2 = filtro[filtro['close'] < 100]
filtro2
```

```python
aapl[aapl['volume']>50000000].loc[aapl['close']<100]
```

otra forma mas prolija

conjuncion de ambas condiciones

```python
volumen_alto = aapl[aapl['volume']>50000000]
precio_menor_a_100 = aapl[aapl['close']<100]
filtrado = pd.merge(left=volumen_alto,right=precio_menor_a_100,how='inner')
filtrado
```

union de ambas, o uno o la otra

```python
outer_aapl = volumen_alto.merge(precio_menor_a_100,how='outer')
outer_aapl
```

filtros avanzados con OPERADORES lógicos

```python
filtro_con_operadores = aapl[(aapl['close']<100) | (aapl['volume']>50000000)]
filtro_con_operadores # equivalente a outer
```

```python
intersecc_con_operadores = aapl[(aapl['close']<100) & (aapl['volume']>50000000)]
intersecc_con_operadores # equivalente a inner
```
