# **Creación de nuevas columnas en un DataFrame**

**Creación de columna con valor constante**

```python
import pandas as pd
```

```python
data=pd.read_excel('AAPL.xlsx', sheet_name='Hoja1')
data.set_index('timestamp', inplace=True)
data['Precio_COME'] = 3
data.drop(['high','low'], axis=1).head()
```

**Creación de columna a partir de otras columnas**

```python
aapl = pd.read_excel('AAPL.xlsx', sheet_name='Hoja1')
aapl.set_index('timestamp', inplace=True)
aapl['precio medio'] = (aapl.open + aapl.close + aapl.low + aapl.high) / 4 
aapl.drop(['high', 'low'], axis=1).head()
```

```python
aapl['vol_mln_usd'] = round((aapl.volume * aapl['precio medio']) / 1000000) # nominales a usd millones
aapl.head()
```

**creación de una columna iterando el df**

```python
AAPL = pd.read_excel('AAPL.xlsx', sheet_name='Hoja1')
AAPL.reset_index()
for i in range(len(AAPL)):
    if AAPL.loc[i, 'close'] >= AAPL.loc[i, 'open']:
        AAPL.loc[i, 'color_vela'] = 'verde'
    else:
        AAPL.loc[i, 'color_vela'] = 'roja'
AAPL.head()
```

**guardado de un df en un excel**

```python
AAPL['mov_intra'] = AAPL.close - AAPL.open
AAPL.to_excel('AAPL_MODIFICADO.xlsx', sheet_name='hoja_ejemplo')
```
